```python
import pandas as pd
import numpy as np


# 1. Create Dataset
data = {
    "Name": ["A", "B", "C", "D", "E"],
    "Math": [78, 90, 65, 88, np.nan],
    "Science": [85, 88, 70, 92, 75],
    "English": [82, 91, 68, 89, 80],
    "City": ["Delhi", "Mumbai", "Delhi", "Chennai", "Mumbai"]
}


df = pd.DataFrame(data)

print("Original Data:\n", df)
df["Math"].fillna(df["Math"].mean(), inplace=True)
df["Total"] = df[["Math", "Science", "English"]].sum(axis=1)
df["Average"] = df[["Math", "Science", "English"]].mean(axis=1)

# 4. Conditional Column

df["Result"] = np.where(df["Average"] >= 80, "Pass", "Fail")


# 5. Sorting

df_sorted = df.sort_values(by="Average", ascending=False)

# 6. Filtering

top_students = df[df["Average"] > 80]
# 7. GroupBy Operation
 
city_avg = df.groupby("City")["Average"].mean()
# 8. Indexing

topper = df.loc[df["Average"].idxmax()]

# 9. Add New Column using NumPy

df["Bonus"] = np.random.randint(1, 6, size=len(df))
df["Final Score"] = df["Total"] + df["Bonus"]


# 10. Rename Col
df.rename(columns={"Math": "Maths Marks"}, inplace=True)


# 11. Drop Column

df.drop(columns=["Bonus"], inplace=True)
#export csv
df.to_csv("students_output.csv", index=False)


# Final Outputs

print("\nFinal Data:\n", df)
print("\nSorted Data:\n", df_sorted)
print("\nTop Students (>80 avg):\n", top_students)
print("\nCity-wise Average:\n", city_avg)
print("\nTopper Details:\n", topper)
```
