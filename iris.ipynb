```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("iris.csv")
```

```python
# Shape of dataset
print("Shape:", df.shape)

# Column names
print("Columns:", df.columns)

# Data types
print(df.dtypes)

print(df['species'].value_counts())
```

```python
sns.barplot(x=df['species'].value_counts().index,
            y=df['species'].value_counts().values)

plt.title("Bar Plot of Species Count")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```

```python
sns.histplot(df['species'],bins=20)
```

```python
sns.boxplot(df['species'])
```

```python
num_cols = df.columns[:-1]
for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True, bins=20)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()

```

```python
for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(data=df, x=col, hue='species', kde=True)
    plt.title(f"{col} distribution by species")
    plt.show()
```

```python
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df)
plt.title("Sepal Length vs Width")
plt.show()

sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=df)
plt.title("Petal Length vs Width")
plt.show()
```

```python
sns.regplot(x='sepal_length', y='sepal_width', data=df)
plt.title("Regression: Sepal Length vs Width")
plt.show()

sns.regplot(x='petal_length', y='petal_width', data=df)
plt.title("Regression: Petal Length vs Width")
plt.show()
```

```python
sns.histplot(df['sepal_length'], kde=True)
plt.show()
```

```python
sns.boxplot(x=df['sepal_width'])
plt.show()
```

```python
sns.histplot(df['petal_length'], kde=True)
plt.show()
```

```python
counts = df['species'].value_counts()

plt.figure(figsize=(5,5))
plt.pie(counts, labels=counts.index, autopct='%1.1f%%')
plt.title("Species Distribution")
plt.show()
```

```python
sns.countplot(x='species', data=df)
plt.title("Countplot of Species")
plt.show()
```

```python
percent = df['species'].value_counts(normalize=True) * 100

sns.barplot(x=percent.index, y=percent.values)
plt.ylabel("Percentage")
plt.title("Species Percentage Distribution")
plt.show()
```

```python
for col in df.columns[:-1]:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], bins=18, kde=True, color='green')
    plt.title(f"Histogram (Bins=15): {col}")
    plt.show()
```
